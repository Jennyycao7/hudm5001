14) **(1 PT)** Write code to generate and plot 10 random walks. For reproducibility, explicitly use seed=314.  
Hint: Calling *plt.plot()* in a *for loop* will plot multiple lines on the same graph. For guidance on using a for loop, refer to the [lecture note on data types](https://github.com/youmilab/hudm5001/blob/main/lecture_notes/python/data_types.ipynb). 

```python
seed(314) #seed
#normal_dist_x = np.random.normal(loc=0, scale=1, size=20)
#normal_dist_y = np.random.normal(loc=0, scale=1, size=20)

for n in range(10):
    x = np.random.normal(loc=0, scale=1, size=100)
    y = np.random.normal(loc=0, scale=1, size=100)
    x_cum = np.cumsum(x)
    y_cum = np.cumsum(y)
    plt.plot(x_cum, y_cum)
    
plt.show()
        

```
