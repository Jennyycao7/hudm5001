```python
import sqlite3
```

2. Creating dataset

```python
emotions = ['grief','overstimulated','emptiness','stress','grief']
frequency = [4,7,3,5,3]
intensity = [10.0,6,4,8.5,5.34]
dataset = list(zip(emotions,frequency,intensity))
```

```python
dataset
```

3. Create table

```python
path_to_db = "/Users/jeffe/sqlite/mydata.db"
```

```python
# create db connection
conn = sqlite3.connect(path_to_db)
```

```python
# create cursor
cur = conn.cursor()
```

```python
#cur.execute('DELETE FROM mytable')
#conn.commit()
```

```python
cur.execute('CREATE TABLE IF NOT EXISTS mytable (emotions TEXT, frequency INTEGAR, intensity REAL)')
conn.commit()
```

```python
cur.executemany('insert into mytable values (?,?,?)', dataset)
conn.commit()
```

```python
# query the table
# print all the rows
for row in conn.execute('select * from mytable'):
    print(row)
```

```python
# print all rows where frequency > 4. show only emotions, intensity
for row in conn.execute('select emotions, intensity from mytable where frequency >4'):
    print(row)
```

```python
query = '''

SELECT
    emotions,
    sum(frequency),
    avg(intensity)
FROM
    mytable
GROUP BY
    emotions;
'''
    
```

```python
cur.execute(query)
```

```python
for row in conn.execute(query):
    print(row)
```

Pandas

```python
# save the resultant dataset in a list
data_p = []
for row in conn.execute(query):
    data_p.append(row)
```
